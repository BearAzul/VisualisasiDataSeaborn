```python
#import library untuk analisis dataa

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from google.colab import files
uploaded = files.upload()
```

```python
df = pd.read_csv('Stocks.csv')
df
```

```python
#a. Distribusi df pada setiap atribut Purchase Price, Current Price, dan Gain/Loss dengan grafik histogram
plt.figure(figsize=(15, 6))

plt.subplot(1, 3, 1)
sns.histplot(df['Purchase Price ($)'], kde=True, bins=10, edgecolor='w', color="k")
plt.xlabel('Purchase Price')
plt.ylabel('Frequency')
plt.title('Histogram dari Purchase Price')

plt.subplot(1, 3, 2)
sns.histplot(df['Current Price ($)'], kde=True, bins=10, edgecolor='k', color="y")
plt.xlabel('Current Price')
plt.ylabel('Frequency')
plt.title('Histogram dari Current Price')

plt.subplot(1, 3, 3)
sns.histplot(df['Gain/Loss ($)'], kde=True, bins=10, edgecolor='w', color="r")
plt.xlabel('Gain/Loss')
plt.ylabel('Frequency')
plt.title('Histogram dari Gain/Loss')

plt.tight_layout()
plt.show()
```

```python
# Sort data by Dividend / Share
data = df.sort_values(by="Dividend/Shares", ascending=False)
# Get the 5th highest Dividend / Share
top_5 = data.head(5)
# Plot the data
plt.figure(figsize=(9, 4))
sns.barplot(x="Stock Name", y="Dividend/Shares", data=top_5)
plt.title("Top 5 Stocks with Highest Dividend / Shares")
plt.show()
```

```python
# Sort data by Dividend / Share
data = df.sort_values(by="Market Value ($)", ascending=False)
# Get the 5th highest Dividend / Share
top_5 = data.head(5)
# Plot the data
plt.figure(figsize=(9, 4))
sns.barplot(x="Stock Name", y="Market Value ($)", data=top_5)
plt.title("Top 5 Stocks with Highest Market Value ($)")
plt.show()
```

```python
# Sort data by Dividend / Share
data = df.sort_values(by="Annual Yield (%)", ascending=False)
# Get the 5th highest Dividend / Share
top_7 = data.head(7)
# Plot the data
plt.figure(figsize=(12, 4))
sns.barplot(x="Stock Name", y="Annual Yield (%)", data=top_7)
plt.title("Top 7 Stocks with Highest Annual Yield (%)")
plt.show()
```
